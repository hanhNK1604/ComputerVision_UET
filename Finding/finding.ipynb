```python
import numpy as np 
import cv2 
import matplotlib.pyplot as plt 
```

```python
image = cv2.imread('image/image.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray, cmap='gray')
```

# TEMPLATE 1

```python
temp1 = cv2.imread('image/temp1.png')
temp1_gray = cv2.cvtColor(temp1, cv2.COLOR_BGR2GRAY)
plt.imshow(temp1_gray, cmap='gray')
```

```python
w, h = temp1_gray.shape[::-1]
res = cv2.matchTemplate(image_gray, temp1_gray, cv2.TM_CCOEFF_NORMED)
loc = np.where(res >= 0.8)

for pt in zip(*loc[::-1]): 
    cv2.rectangle(image_rgb, pt, (pt[0] + w, pt[1] + h), (0, 0, 255), 2)

plt.imshow(image_rgb)
    
```
